```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
quality_counts = df['popularity'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color='deeppink')
plt.title('Distribution of popularity')
plt.xlabel('popularity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
plt.boxplot(df['popularity'].dropna(), vert=False)
plt.xlabel("Popularity")
plt.title("Boxplot of Track Popularity")
plt.tight_layout()
plt.show()
```

```python
genre_pop = df.groupby("track_genre")["popularity"].agg(["mean", "count"]).sort_values("mean", ascending=False)

genre_pop.head(10)
```

```python
audio_cols = [
    "danceability", "energy", "loudness", "speechiness", "acousticness",
    "instrumentalness", "liveness", "valence", "tempo", "duration_ms"
]

corr = df[audio_cols + ["popularity"]].corr()
print(corr["popularity"].sort_values(ascending=False))
```
